```python
import shlex, subprocess
from pathlib import Path

PROJECT_ROOT = Path.cwd()
DATA_ROOT = (PROJECT_ROOT / '../data').resolve()
print(f'工作目录: {PROJECT_ROOT}')
print(f'数据目录: {DATA_ROOT}')

# 按需修改下方参数，即可在 Colab 上复现不同实验设置
train_params = {
    'model_name': 'SASRec',
    'model_mode': '',
    'dataset': 'MovieLens_1M',
    'path': str(DATA_ROOT),
    'gpu': '0',
    'random_seed': 2024,
    'history_max': 50,
    'num_layers': 2,
    'num_heads': 2,
    'emb_size': 128,
    'dropout': 0.1,
    'lr': 1e-4,
    'l2': 1e-6,
    'batch_size': 512,
    'eval_batch_size': 1024,
    'num_workers': 4,
    'metric': 'NDCG,HR',
    'topk': '5,10,20',
    'main_metric': 'NDCG@5',
    'epoch': 100,
    'early_stop': 10,
    'test_epoch': 1,
    'train': 1,
    'buffer': 1
}

cmd = ['python', 'main.py']
for key, value in train_params.items():
    if value is None or value == '':
        continue
    cmd.extend([f'--{key}', str(value)])

print('运行命令:\n', ' '.join(shlex.quote(c) for c in cmd))
subprocess.run(cmd, check=False)
```
